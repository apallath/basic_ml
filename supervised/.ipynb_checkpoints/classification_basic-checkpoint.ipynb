```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
%matplotlib inline
```

# k-NN classification using the IRIS dataset

```python
iris = pd.read_csv('iris.data')
iris.head()
#add class labels
d = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
iris['class_label'] = iris.apply(lambda row: d[row['class']], axis = 1)
iris
```

```python
#complete dataset
X = iris[['sepal length', 'sepal width', 'petal length', 'petal width']]
y = iris[['class_label']]
#split dataset into training set and test set, randomly selecting data. By default - 75-25 split
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y,random_state = 5) #seed is 5 
print X.shape, X_train.shape, X_test.shape
print X_train.head()
print y_train.head()
```

```python
from matplotlib import cm
scatter = pd.plotting.scatter_matrix(X_train, c = y_train, marker = 'o', s = 40, figsize=(18,18))
```

```python
from sklearn.neighbors import KNeighborsClassifier

uniform_knn = KNeighborsClassifier(n_neighbors = 5, weights = 'uniform')
distance_knn = KNeighborsClassifier(n_neighbors = 5, weights = 'distance')

uniform_knn.fit(X_train, y_train.values.ravel())  #ravel returns a flattened numpy array
distance_knn.fit(X_train, y_train.values.ravel())
```

```python
print uniform_knn.score(X_test, y_test)
print distance_knn.score(X_test, y_test)
```

# k-NN classification using scikit-learn generated dataset

```python
from sklearn.datasets import make_classification
#2 features, 4 classes
[X,y] = sklearn.datasets.make_classification(n_samples = 300, n_features = 2, 
                                     n_informative = 2, n_redundant = 0, n_repeated = 0,
                                     n_classes = 2)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)
print X_train.shape, y_train.shape
print X_test.shape, y_test.shape
```

```python
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(X_train, y_train)
print knn.score(X_train,y_train) #training score
print knn.score(X_test,y_test) #test score
```

```python
#plot decision boundaries

#x and y axis min and max
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1

#select mesh step-size
h = 0.02
#prepare mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

#concatenate along second axis to generate ndarray with point pairs
points = np.c_[xx.ravel(), yy.ravel()]

#predict for each point in the meshgrid
Z = knn.predict(points)
Z = Z.reshape(xx.shape)

#plot
from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])

plt.figure(figsize=(10,10))
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, s=40)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```
